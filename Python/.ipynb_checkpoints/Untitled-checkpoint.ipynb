```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import urllib
import os 
import requests 
import pandas as pd 

servico = Service(ChromeDriverManager().install())
nav = webdriver.Chrome(service=servico)

diretorio_destino = r"C:\Users\Mateus\Desktop\projeto-listagem-pokemon\src\imagens"
for i in range(1,152):
    nav.get(f"https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/versions/generation-v/black-white/animated/{i}.gif")
    img = nav.find_element(By.XPATH,'/html/body/img')
    src = img.get_attribute("src")
    nome_arquivo = f"foto{i}.gif"
    caminho_completo = os.path.join(diretorio_destino, nome_arquivo)
    urllib.request.urlretrieve(src, caminho_completo)
```

```python
def obter_descricoes_e_nomes_kanto():
    url = 'https://pokeapi.co/api/v2/pokedex/kanto'
    pokemon_dict = {}
    
    r = requests.get(url)
    
    if r.status_code == 200:
        data = r.json()
        for entry in data['pokemon_entries']:
            pokemon_url = entry['pokemon_species']['url']
            pokemon_data = requests.get(pokemon_url).json()
            
            nome = pokemon_data['name']
            descricao = next((flavor['flavor_text'] for flavor in pokemon_data['flavor_text_entries'] 
                              if flavor['language']['name'] == 'en'), '')
            pokemon_dict[nome] = descricao
            
    else:
        print("Não foi possível obter os dados dos Pokémon.")
    
    return pokemon_dict

kanto_pokemon_dict = obter_descricoes_e_nomes_kanto()

for nome, descricao in kanto_pokemon_dict.items():
    print(f"Nome: {nome}")
    print(f"Descrição: {descricao}\n")
```

```python
def obter_tipos_pokemon_kanto():
    df = pd.read_csv('pokemon.csv')

    tipos_pokemon = {}

    for _, row in df.iterrows():
        if row['generation'] == 1:
            nome = row['name']
            tipo1 = row['type1']
            tipo2 = row['type2'] if not pd.isnull(row['type2']) else None

            tipos = [tipo1]
            if tipo2:
                tipos.append(tipo2)

            tipos_pokemon[nome] = tipos

    return tipos_pokemon

tipos_pokemon_kanto = obter_tipos_pokemon_kanto()

if tipos_pokemon_kanto:
    print("Tipos dos Pokémon de Kanto:")
    for nome, tipos in tipos_pokemon_kanto.items():
        print(f"{nome}: {tipos}")
else:
    print("Falha ao obter os tipos dos Pokémon de Kanto.")
```

```python
html_body = """
    <li class="pokemon-card">
        <div class="informations">
            <span>{nome_pokemon}</span>
            <span>#{numero}</span>
        </div>

        <img src="./src/imagens/foto{i}.gif" alt="{nome_pokemon}" class="gif">

        <ul class="types">
            {tipos_html}
        </ul>

        <p class="description">{descricao}</p>
    </li>
"""
for i, (nome, descricao) in enumerate(kanto_pokemon_dict.items(), start=1):
    tipos = tipos_pokemon_kanto.get(nome.capitalize(), [])
    tipos_html = ''.join([f'<li class="type {tipo.lower()}">{tipo.capitalize()}</li>' for tipo in tipos])
    
    if tipos_html:
        tipos_html = '\n' + tipos_html  
    
    
    numero = str(i).zfill(3)
    
    pokemon_html = html_body.format(nome_pokemon=nome.capitalize(), numero=numero, tipos_html=tipos_html, descricao=descricao,i=i) 
    print(pokemon_html)
```
